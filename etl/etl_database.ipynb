```python
from prefect import flow, task
import pandas as pd
import sqlite3
import os
from pathlib import Path

DB_PATH = "../stage/world_data.db"
STAGE_DIR = Path("../stage")


@task
def reset_db():
    """
    Reinicia la base borrando el archivo SQLite si existe.
    """
    if os.path.exists(DB_PATH):
        os.remove(DB_PATH)


@task
def load_csv_dynamic(table_name: str, file_path: str):
    """
    Carga un CSV en SQLite.
    - Si la tabla ya existe, la reemplaza.
    - Si no existe, la crea automáticamente con el esquema del CSV.
    """
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"No existe el archivo: {file_path}")

    conn = sqlite3.connect(DB_PATH)
    df = pd.read_csv(file_path)

    # pandas infiere el esquema → no se rompe si cambian las columnas
    df.to_sql(table_name, conn, if_exists="replace", index=False)

    conn.commit()
    conn.close()
    print(f"Tabla '{table_name}' cargada desde {file_path}")


@flow(name="etl_sqlite_dynamic_flow")
def etl_sqlite_dynamic_flow():
    reset_db()

    # Buscar todos los CSV en ../stage
    csv_files = STAGE_DIR.glob("*.csv")

    for file_path in csv_files:
        table_name = file_path.stem  # nombre de archivo sin extensión
        load_csv_dynamic(table_name, str(file_path))


if __name__ == "__main__":
    etl_sqlite_dynamic_flow()
```
