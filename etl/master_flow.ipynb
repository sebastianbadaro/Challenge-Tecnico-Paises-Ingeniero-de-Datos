```python
# master_flow.py
from prefect import flow
import papermill as pm
from pathlib import Path
etl_dir = Path.cwd()

import os
os.environ["DISABLE_PANDERA_IMPORT_WARNING"] = "True"

def run_notebook(path: Path):
    try:
        print(f"▶ Ejecutando {path.name} ...")
        pm.execute_notebook(
            str(path),
            "NUL",   # salida descartada
            log_output=True
        )
        print(f"✅ {path.name} ejecutado con éxito")
    except Exception as e:
        print(f"⚠ Error en {path.name}: {e}")

@flow(name="world_data_master_flow")
def world_data_master_flow():
    # Detectar todos los notebooks ETL
    notebooks = sorted(etl_dir.glob("etl_*.ipynb"))

    # Forzar orden: Country primero, Database último
    first = [etl_dir / "etl_Country.ipynb"]
    last = [etl_dir / "etl_Database.ipynb"]

    # Filtrar el resto (ni Country ni Database)
    middle = [nb for nb in notebooks if nb not in first + last]

    # Orden final
    ordered = first + middle + last

    # Ejecutar en orden
    for nb in ordered:
        if nb.exists():
            run_notebook(nb)
        else:
            print(f"⚠ Saltado: {nb.name} no encontrado")

if __name__ == "__main__":
    world_data_master_flow()
```
